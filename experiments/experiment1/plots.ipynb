```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data.csv', delimiter='\t')
```

```python
print('Kgat precision loss (from 10 to 30)', 0.1359 / 0.1978 * 100)
print('CKE precision loss (from 10 to 30)', 0.1352 / 0.1959 * 100)
print('KGCN precision loss (from 10 to 30)', float(df[df['model'] == 'KGCN'][df['metric'] == 'precision'][df['k'] == 30]['value']) / float(df[df['model'] == 'KGCN'][df['metric'] == 'precision'][df['k'] == 10]['value']) * 100)
print('MKR precision loss (from 10 to 30)', float(df[df['model'] == 'MKR'][df['metric'] == 'precision'][df['k'] == 30]['value']) / float(df[df['model'] == 'MKR'][df['metric'] == 'precision'][df['k'] == 10]['value']) * 100)
print('RippleNET precision loss (from 10 to 30)', float(df[df['model'] == 'RippleNET'][df['metric'] == 'precision'][df['k'] == 30]['value']) / float(df[df['model'] == 'RippleNET'][df['metric'] == 'precision'][df['k'] == 10]['value']) * 100)
print('ItemKNN precision loss (from 10 to 30)', float(df[df['model'] == 'ItemKNN'][df['metric'] == 'precision'][df['k'] == 30]['value']) / float(df[df['model'] == 'ItemKNN'][df['metric'] == 'precision'][df['k'] == 10]['value']) * 100)
print('BPR precision loss (from 10 to 30)', float(df[df['model'] == 'BPR'][df['metric'] == 'precision'][df['k'] == 30]['value']) / float(df[df['model'] == 'BPR'][df['metric'] == 'precision'][df['k'] == 10]['value']) * 100)
```

```python
metrics = ['recall', 'ndcg', 'precision']
# metric = 'map'

for metric in metrics:
  fig, axs = plt.subplots(1, 1,  figsize=([8, 5]))
  axs.plot(df[df['model'] == 'KGAT'][df['metric'] == metric]['k'] , df[df['model'] == 'KGAT'][df['metric'] == metric]['value'], marker='x', label='KGAT', )
  axs.plot(df[df['model'] == 'MKR'][df['metric'] == metric]['k'] , df[df['model'] == 'MKR'][df['metric'] == metric]['value'], marker='v', label='MKR')
  axs.plot(df[df['model'] == 'KGCN'][df['metric'] == metric]['k'] , df[df['model'] == 'KGCN'][df['metric'] == metric]['value'], marker='o', label='KGCN')
  axs.plot(df[df['model'] == 'CKE'][df['metric'] == metric]['k'] , df[df['model'] == 'CKE'][df['metric'] == metric]['value'], marker='s', label='CKE')
  axs.plot(df[df['model'] == 'RippleNET'][df['metric'] == metric]['k'] , df[df['model'] == 'RippleNET'][df['metric'] == metric]['value'], marker='*', label='RippleNET')
  axs.plot(df[df['model'] == 'ItemKNN'][df['metric'] == metric]['k'] , df[df['model'] == 'ItemKNN'][df['metric'] == metric]['value'], marker='+', label='ItemKNN')
  axs.plot(df[df['model'] == 'BPR'][df['metric'] == metric]['k'] , df[df['model'] == 'BPR'][df['metric'] == metric]['value'], marker='d', label='BPR')


  plt.xticks(np.arange(10, 51, 10))
  axs.set_xlabel('K')
  axs.set_ylabel(metric)
  axs.legend(bbox_to_anchor=[1, 1],
                  ncol=1, title="Model name")

  fig.savefig(f"./plots2/{metric}.png", bbox_inches='tight')
```
