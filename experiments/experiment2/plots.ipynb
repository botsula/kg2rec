```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
df = pd.read_csv('./data.csv', delimiter='\t')
```

```python
fig, axs = plt.subplots(1, 1,  figsize=([15, 10]))
axs.plot(df[df['model'] == 'KGAT']['train_%'] , df[df['model'] == 'KGAT']['ndcg@20'], marker='x', label='KGAT', )
axs.plot(df[df['model'] == 'MKR']['train_%'] , df[df['model'] == 'MKR']['ndcg@20'], marker='v', label='MKR')
axs.plot(df[df['model'] == 'KGCN']['train_%'] , df[df['model'] == 'KGCN']['ndcg@20'], marker='o', label='KGCN')
axs.plot(df[df['model'] == 'CKE']['train_%'] , df[df['model'] == 'CKE']['ndcg@20'], marker='s', label='CKE')
axs.plot(df[df['model'] == 'RippleNet']['train_%'] , df[df['model'] == 'RippleNet']['ndcg@20'], marker='*', label='RippleNET')
axs.plot(df[df['model'] == 'ItemKNN']['train_%'] , df[df['model'] == 'ItemKNN']['ndcg@20'], marker='+', label='ItemKNN')
axs.plot(df[df['model'] == 'BPR']['train_%'] , df[df['model'] == 'BPR']['ndcg@20'], marker='d', label='BPR')


plt.xticks(np.arange(30, 91, 30))
axs.set_ylabel('ndcg@20')
axs.set_xlabel('Train set %')
axs.legend(bbox_to_anchor=[1, 1],
                ncol=1, title="Legend")
axs.grid(linestyle='-', linewidth=0.5)

fig.savefig(f"./plots/sparsity_test.png", bbox_inches='tight')
```

```python
float(df[df['model'] == 'KGCN'][df['train_%'] == 30]['ndcg@20']) / float(df[df['model'] == 'KGCN'][df['train_%'] == 60]['ndcg@20']) * 100
```

```python
for model in ['KGCN', 'BPR', 'CKE', 'ItemKNN', 'KGAT', 'MKR', 'RippleNet']:
  res = float(df[df['model'] == model][df['train_%'] == 60]['ndcg@20']) / float(df[df['model'] == model][df['train_%'] == 30]['ndcg@20']) * 100
  print(model, '&', round(res, 2), '\% \\\ ')
```

```python
train = 90

fig, ax = plt.subplots(figsize=([8, 5]))
models = [ 'KGAT', 'MKR', 'KGCN', 'CKE',  'RippleNet', 'ItemKNN', 'BPR']
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink']

x_index = [0,1,2,3,4,5,6]
y_set = []

for j in x_index:
       y = df[df['train_%'] == train][df['model'] == models[j]]['ndcg@20']
       y_set.append(float(y))

sorted_models = [x for _, x in sorted(zip(y_set, models))]
sorted_colors = [x for _, x in sorted(zip(y_set, colors))]
y_set.sort()

# ax.bar(x_index, y_set, width=1, edgecolor="white", linewidth=0.7)
ax.bar(x_index, y_set, width=1, edgecolor="white", linewidth=0.7, color=sorted_colors)

x_index = [0,1,2,3,4,5,6]

# plt.xticks(x_index, models)
plt.xticks(x_index, sorted_models)
ax.set_ylabel('ndcg@20')
ax.set_xlabel('Model name')
ax.set_title(f'Train dataset part - {train}%')

ax.set(ylim=(0, 0.32))

plt.show()
fig.savefig(f"./plots2/sparsity_test_{train}.png", bbox_inches='tight')
```
